```python
raw_hospital_data_df.info()
```

From the initial output of the dataframe we can see that the dataframe references null values as "?". Pandas does not recognise this and so this value has to be replaced in the dataframe. To do this we will replace "?" with "NaN".

```python
raw_hospital_data_df = raw_hospital_data_df.replace("?",np.nan)
raw_hospital_data_df
```
